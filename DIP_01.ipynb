# Function

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def show_mnist(mnist_data, i):
  a1 = mnist_data.values[i,1:]   #print(a1) 
  a1_label = mnist_data.values[i,0] #print(a1_label)
  a1_re = np.reshape(a1, (28,28))
    
  plt.gray()
  plt.axis('off')
  #plt.xlabel(a1_re.shape[0])
  #plt.ylabel(a1_re.shape[1])  
  plt.title(a1_label)
  plt.imshow(a1_re)
  #plt.show()
```

```
mnist_data = pd.read_csv('sample_data/mnist_train_small.csv')
#calling function
show_mnist(mnist_data, 0)
```

```
for i in range(0, 5):
  #calling function
  show_mnist(mnist_data, i)
```

```
import matplotlib.pyplot as plt
import random
import numpy as np

r = 3
c = 5

fig = plt.figure()
fig.suptitle('MNIST dataset')

for i in range(1, r*c+1):
  plt.subplot(r, c, i) 
  show_mnist(mnist_data, i)
plt.show()
```

#แสดงเฉพาะบางตัวเลข

```
r= 2
c=5

for n in range(0,10):
  mnist_no = mnist_data.loc[mnist_data['6'] == n]
  fig = plt.figure()
  fig.suptitle(n, fontsize=20)
  
  for i in range(1, r*c+1):
    plt.subplot(r, c, i) 
    show_mnist(mnist_no, i)
  plt.show()
  print()
```

```
mnist_1 = mnist_data.loc[mnist_data['6'] == 1]

for i in range(1, 12):
  plt.subplot(3, 3, i) 
  show_mnist(mnist_1, i)
  
plt.show()
```

```
mnist_1 = mnist_data.loc[mnist_data['6'] == 1]

for i in range(1, 12):
  try:
    plt.subplot(3, 3, i) 
    show_mnist(mnist_1, i)
  except:
    pass
    
plt.show()
```

```
import math

mnist_1 = mnist_data.loc[mnist_data['6'] == 1]

show_num = 17
rc = math.floor(math.sqrt(show_num))

for i in range(1, show_num):
  try:
    plt.subplot(rc, rc, i) 
    show_mnist(mnist_1, i)
  except:
    pass
    
plt.show()
```

```
import cv2
import matplotlib.pyplot as plt

def cv_read_image(path):
  img = cv2.imread(path)
  img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  return img_rgb

r = 1
c = 2
path = ['dog.jpg', 'dog1.jpg']

cnt=0
plt.figure(figsize=(15,10))
for i in range(1, len(path)+1):
  try:
    plt.subplot(r, c, i) 
    img = cv_read_image(path[cnt])  
    plt.imshow(img)
    cnt = cnt+1
  except:
    pass
      
plt.show()
```
