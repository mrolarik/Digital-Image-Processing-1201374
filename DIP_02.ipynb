```
path = 'sample_text.png'
img = cv2.imread(path)
#img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.title('Gray image')
plt.imshow(img_gray)
plt.show()

(thresh, img_bw) = cv2.threshold(img_gray, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print('Threshold value', thresh)

plt.title('Black and white image')
plt.imshow(img_bw)
plt.show()
```

```
import numpy as np

blurred = cv2.GaussianBlur(img_gray, (3, 3), 0)
canny = cv2.Canny(blurred, 120, 255, 1)
kernel = np.ones((5,5),np.uint8)
dilate = cv2.dilate(canny, kernel, iterations=1)

plt.title('Blurred')
plt.imshow(blurred)
plt.show()

plt.title('Canny')
plt.imshow(canny)
plt.show()

plt.title('Dilate')
plt.imshow(dilate)
plt.show()
```

```
# Find contours
cnts = cv2.findContours(dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cnts = cnts[0] if len(cnts) == 2 else cnts[1]

print('Number of contours', len(cnts))
```

```
# copy array
img_detect = np.array(img, copy=True)
```

```
# Iterate thorugh contours and filter for ROI
image_number = 0
for c in cnts:
    x,y,w,h = cv2.boundingRect(c)
    cv2.rectangle(img_detect, (x, y), (x + w, y + h), (36,255,12), 2)
    #ROI = img[y:y+h, x:x+w]
    #cv2.imwrite("ROI_{}.png".format(image_number), ROI)
    image_number += 1
```

```
plt.title('Object Detection')
plt.imshow(img_detect)
plt.show()
```

```
import numpy as np

def text_detection(path, thresh=127):
  #path = 'sample_text.png'
  img = cv2.imread(path)
  img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  
  #(thresh, img_bw) = cv2.threshold(img_gray, 128, 255, cv2.THRESH_BINARY 
              #| cv2.THRESH_OTSU)  
  img_bw = cv2.threshold(img_gray, thresh, 255, cv2.THRESH_BINARY)[1]
  #print('Threshold value', thresh)

  # invert image  
  img_bw = cv2.bitwise_not(img_bw)
  
  # Find contours
  cnts = cv2.findContours(img_bw, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
  cnts = cnts[0] if len(cnts) == 2 else cnts[1]
  print('Number of contours', len(cnts))
  
  # copy array
  img_detect = np.array(img, copy=True)

  # Iterate thorugh contours and filter for ROI
  image_number = 0
  for c in cnts:
    x,y,w,h = cv2.boundingRect(c)
    cv2.rectangle(img_detect, (x, y), (x + w, y + h), (36,255,12), 2)
    image_number += 1

  return img_detect 
```

```
path = 'sample_text.png'

img_text = text_detection(path, thresh=127)
plt.imshow(img_text)
plt.show()

img_text = text_detection(path)
plt.imshow(img_text)
plt.show()
```
